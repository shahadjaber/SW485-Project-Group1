# Online Purchase Prediction System

## Introduction & Problem Statement

Online businesses often struggle to determine whether a website visitor will complete a purchase. The objective of this project is to build a machine learning classification model that predicts whether a user session will result in a purchase (Revenue = TRUE) or not (Revenue = FALSE).

## Dataset Selection & Justification

The dataset used in this project is the Online Shoppers Intention Dataset obtained from Kaggle:

https://www.kaggle.com/datasets/henrysue/online-shoppers-intention

The dataset contains over 12,000 user sessions and 18 features describing browsing behavior. The target variable is “Revenue”, a binary classification label indicating whether a purchase was made.

This dataset is suitable because:
- It is structured tabular data.
- It supports binary classification.
- It contains sufficient observations.
- It represents a real-world business problem.
